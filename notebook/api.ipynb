```python
from meetup.mvg_5 import MyJob
import pandas as pd
import numpy as np
import datetime
```

```python
data = MyJob().cookie.get("departures")
len(data)
```

```python
df = pd.DataFrame(
    data, columns=[
        "station", 
        "product", 
        "label", 
        "color", 
        "destination", 
        "departure", 
        "distance", 
        "name"
    ]
)
df
```

```python
df["walk"] = df.apply(
    lambda r: (
        r["departure"] 
        - datetime.timedelta(
            minutes=np.ceil(r["distance"] / 1000. / 3.5 * 60.))
    ),
    axis=1
)
```

```python
df.head()
```

```python
df.sort_values("walk", inplace=True)
g = df.groupby([
    "station", "product", "label", "destination"]).first()
df = pd.DataFrame(g)
df.reset_index(inplace=True)
df.sort_values("walk", inplace=True)
df
```
